## List Comprehension

```python
stop_words = ['I', 'love', 'and', 'to']
sentence = 'I love AI and listen to music'

new_sentence = [word for word in sentence.split(' ') if word not in stop_words]
new_sentence
```

## Tuple

```python
import numpy as np
import math
```

```python
my_tuple1 = (2,3)
my_tuple2 = (3,6)

def new_vector(vector):
    s = 0
    mul = 1
    for value in vector:
        s += value
        mul *= value
    return s, mul

dist = math.sqrt(sum((i-j)**2 for i, j in zip(my_tuple1,my_tuple2)))
index = (my_tuple1.index(3),my_tuple2.index(3))

print('Kết quả vector 1 :', new_vector(my_tuple1))
print('Kết quả vector 2 :', new_vector(my_tuple2))
print('Khoảng cách: ', dist)
print('Vị trí giá trị 3: ', index)
```

```python
my_tuple1 = (2,3)
my_tuple2 = (3,6)

tong_vector1 = (my_tuple1[0] + my_tuple2[0], my_tuple1[1] + my_tuple2[1])
print('Tổng 1 của 2 tuple:', tong_vector1)

tong_vector2 = tuple(list(map(lambda x, y : x + y, my_tuple1, my_tuple2)))
print('Tổng 2:của 2 tuple', tong_vector2)
```

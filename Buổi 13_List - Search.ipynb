## Buổi 13 - Search

```python
lst7 = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]

for index, value in enumerate(lst7):
    if value == None:
        break

print('Giá trị None ở vị trí', index)
```

```python
lst_index_None = [index for index, value in enumerate(lst7) if value == None]
print('Giá trị None ở các vị trí', lst_index_None)
```

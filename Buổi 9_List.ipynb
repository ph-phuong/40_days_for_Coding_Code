```python
#Cách 2
def cal_median(lst):
    n = len(lst)
    sorted_lst = sorted(lst)
    if n % 2 == 0:
        return (sorted_lst[n//2] + sorted_lst[n//2-1]) / 2
    else:
        return sorted_lst[n//2]
    
median1 = cal_median(lst)
print(f'Giá trị median là : {median1}')
```

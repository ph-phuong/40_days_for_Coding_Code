```python
import math
```

```python
def quadratic_equation(a, b, c):
    delta = b*b - 4*a*c
    
    if a == 0:
        if b != 0 :
            print('Phng trnh c nghim duy nht : ', -c/b)
        else:
            if c != 0:
                print('Phng trnh v nghim')
            else:
                print('Phng trnh v s nghim')
    else:
        if delta == 0:
            print('Phng trnh c nghim kp :', -b/2*a)
        elif delta < 0:
            print('Phng trnh v nghim')
        else:
            x1 = (-b + math.sqrt(delta)) / (2*a)
            x2 = (-b - math.sqrt(delta)) / (2*a)
            print(f'Phng trnh c 2 nghim x1 = {x1}, x2 = {x2}')

quadratic_equation(a=2, b=6, c=4)
quadratic_equation(a=1, b=2, c=1)
quadratic_equation(a=4, b=6, c=3)
```

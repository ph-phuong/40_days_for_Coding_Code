```python
import math
```

```python
def quadratic_equation(a, b, c):
    delta = b*b - 4*a*c
    
    if a == 0:
        if b != 0 :
            print('Phương trình có nghiệm duy nhất : ', -c/b)
        else:
            if c != 0:
                print('Phương trình vô nghiệm')
            else:
                print('Phương trình vô số nghiệm')
    else:
        if delta == 0:
            print('Phương trình có nghiệm kép :', -b/2*a)
        elif delta < 0:
            print('Phương trình vô nghiệm')
        else:
            x1 = (-b + math.sqrt(delta)) / (2*a)
            x2 = (-b - math.sqrt(delta)) / (2*a)
            print(f'Phương trình có 2 nghiệm x1 = {x1}, x2 = {x2}')

quadratic_equation(a=2, b=6, c=4)
quadratic_equation(a=1, b=2, c=1)
quadratic_equation(a=4, b=6, c=3)
```

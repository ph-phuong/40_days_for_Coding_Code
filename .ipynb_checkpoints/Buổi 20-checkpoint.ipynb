```python
from newspaper import Article
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
```

```python
article = Article('https://vnexpress.net/tran-mua-lon-nhat-75-nam-khien-dubai-ngap-lut-4735580.html')

article.download()
article.parse()

print(article.text)
print(article.images)
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

data = article.text
tokenization = word_tokenize(data)
result = pos_tag(tokenization)
print(result)
```

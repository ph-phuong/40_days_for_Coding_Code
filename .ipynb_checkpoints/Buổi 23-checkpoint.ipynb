```python
import numpy as np
import tensorflow as tf
import torch
```

```python
# Câu 1
# Numpy code
np.random.seed(2024)
arr_np = np.arange(0, 10, step = 1)
print('Numpy code: ', arr_np)

# Pytorch code
torch.manual_seed(2024)
arr_torch = torch.arange(0,10, step = 1)
print('Pytorch code: ', arr_torch)

# Tensorflow code
tf.random.set_seed(2024)
arr_tf = tf.range(0, 10, delta=1)
print('Tensorflow code:', arr_tf)
```

```python
# Câu 2
# Numpy code
np.random.seed(2024)
arr_matrix_np = np.eye(3,3)
print('Numpy code: \n', arr_matrix_np)

# Pytorch code
torch.manual_seed(2024)
arr_matrix_torch = torch.eye(3, 3)
print('Pytorch code: \n', arr_matrix_torch)

# Tensorflow code
tf.random.set_seed(2024)
arr_matrix_tf = tf.eye(3, 3)
print('Tensorflow code: \n', arr_matrix_tf)
```

```python
# Câu 3
# Numpy code
np.random.seed(2024)
arr_random_rand_np = np.random.rand(3,4)
arr_random_randint_np = np.random.randint(-10, 10, size = (3,4))
print('Numpy code_rand: \n', arr_random_rand_np)
print('Numpy code_randint: \n', arr_random_randint_np)

# Pytorch code
torch.manual_seed(2024)
arr_random_rand_torch = torch.rand(3,4)
arr_random_randint_torch = torch.randint(-10, 10, size = (3, 4))
print('Pytorch code_rand: \n', arr_random_rand_torch)
print('Pytorch code_randint: \n', arr_random_randint_torch)

# Tensorflow code
tf.random.set_seed(2024)
arr_random_uniform1_tf = tf.random.uniform(shape = (3, 4))
arr_random_uniform2_tf = tf.random.uniform(shape = (3, 4), minval = -10, maxval = 10, dtype =tf. dtypes . int32)
print('Tensorflow code_uniform1: \n', arr_random_uniform1_tf)
print('Tensorflow code_uniform2: \n', arr_random_uniform2_tf)
```

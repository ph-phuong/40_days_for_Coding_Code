```python
import underthesea as ut
```

```python
sentence = '''Cng c Suno AI nhanh chng nhn c s ch  t ngi dng khi c th
            to bi ht ch vi vi cu lnh. Phin bn mi nht V3 Alpha mi c gii
            thiu cui thng 2, c bn min ph vi 10 bi ht mi ngy.'''
```

```python
print('Pos Tag: ', ut.pos_tag(sentence))
print('-------')
print('Classify: ', ut.classify(sentence))
print('-------')
print('Sentiment: ', ut.sentiment(sentence))
print('-------')
print('Sent_tokenize: ', ut.sent_tokenize(sentence))
print('-------')
print('Word tokenize: ', ut.word_tokenize(sentence))
```

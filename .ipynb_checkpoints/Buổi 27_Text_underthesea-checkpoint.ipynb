```python
import underthesea as ut
```

```python
sentence = '''Công cụ Suno AI nhanh chóng nhận được sự chú ý từ người dùng khi có thể
            tạo bài hát chỉ với vài câu lệnh. Phiên bản mới nhất V3 Alpha mới được giới
            thiệu cuối tháng 2, có bản miễn phí với 10 bài hát mỗi ngày.'''
```

```python
print('Pos Tag: ', ut.pos_tag(sentence))
print('-------')
print('Classify: ', ut.classify(sentence))
print('-------')
print('Sentiment: ', ut.sentiment(sentence))
print('-------')
print('Sent_tokenize: ', ut.sent_tokenize(sentence))
print('-------')
print('Word tokenize: ', ut.word_tokenize(sentence))
```

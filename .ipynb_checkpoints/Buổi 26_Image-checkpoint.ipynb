```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Đọc và hiển thị ảnh
img1 = cv2.imread("Buoi_26-1.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
plt.axis('off')
plt.show()
```

```python
# Đọc và hiển thị ảnh màu xám
img2 = cv2.imread("Buoi_26-1.jpg", 0)
plt.imshow(img2, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Tăng cường độ sáng ảnh lên 50 đơn vị
new_img1 = img1.astype(np.float32) + 50
new_img1 = np.clip(new_img1, 0, 255)
new_img1 = new_img1.astype(np.uint8)

plt.imshow(new_img1)
plt.axis('off')
plt.show()
```

```python
# Giảm cường độ sáng ảnh lên 80 đơn vị
new_img2 = img1.astype(np.float32) - 80
new_img2 = np.clip(new_img2, 0, 255)
new_img2 = new_img2.astype(np.uint8)

plt.imshow(new_img2)
plt.axis('off')
plt.show()
```

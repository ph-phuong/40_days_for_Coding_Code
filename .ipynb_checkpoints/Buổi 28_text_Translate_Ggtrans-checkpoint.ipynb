```python
import translate as trans
import googletrans as ggtrans
```

```python
print(ggtrans.LANGUAGES)
```

```python
sentence = '''Công cụ Suno AI nhanh chóng nhận được sự chú ý từ người dùng khi có thể
            tạo bài hát chỉ với vài câu lệnh. Phiên bản mới nhất V3 Alpha mới được giới
            thiệu cuối tháng 2, có bản miễn phí với 10 bài hát mỗi ngày.'''
```

```python
# translate

translator_en_trans = trans.Translator(from_lang = 'vi', to_lang ="en")
translation_en_trans = translator_en_trans.translate(sentence)
print('Dịch tiếng Anh bằng translate: \n', translation_en_trans)
print('-------')
translator_ja_trans = trans.Translator(to_lang ="ja")
translation_ja_trans = translator_ja_trans.translate (sentence)
print('Dịch tiếng Nhật bằng translate: \n', translation_ja_trans)
```

```python
# googletran
translator_en_ggtrans = ggtrans.Translator()
translation_en_ggtrans = translator_en_ggtrans.translate(sentence,dest = 'en')
print('Dịch tiếng Anh bằng Googletrans: \n', translation_en_ggtrans)
print('-------------')
translator_ja_ggtrans = ggtrans.Translator()
translation_ja_ggtrans = translator_ja_ggtrans.translate(sentence,dest = 'ja')
print('Dịch tiếng Nhật bằng Googletrans: \n', translation_ja_ggtrans)
```

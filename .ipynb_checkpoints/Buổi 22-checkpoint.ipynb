## Vòng lặp While

Đề bài này có thể hiểu là :
- Tạo 2 số ngẫu nhiên a, b chạy trong khoảng (1,20) 
- Tổng của chúng là 1 số nào đó (khác total)
- Số tổng này (giá trị khác total) sẽ được tạo bao nhiêu lần (từ 2 số ngẫu nhiên a, b) cho đến khi = total

```python
import random
def random_number_with_condition(total):
    random.seed(0)
    times = 0 
    a  = b = 0
    while (a + b) != total:
        a = random.randint(1, 20)
        b = random.randint(1, 20)
        times += 1
    print(times)

random_number_with_condition(40)
random_number_with_condition(20)
random_number_with_condition(35)
```

## Vòng lặp while, continue và break

```python
def find_divisible_number(a):
    i = 100
    while True:
        i += 1
        if i % a != 0:
            continue
        else:
            print(i)
        break
    
    
find_divisible_number(5)
find_divisible_number(17)
```

```python
def find_divisible_number2(a):
    i = 101
    while i % a != 0:
        i += 1
    print(i)
        
find_divisible_number2(5)
find_divisible_number2(17)
```

## Vòng lặp while - Newton's method

```python
def find_squared_root(a):
    x = a
    e = 0.001
    while True:
        if abs( -(x**2 - a) / (x*2) ) > e:  
            x = x -(x**2 - a) / (x*2)       # phương pháp Newton: x = x - (x**2 - a) / (2*x).
        else: 
            break
    return x
    
    
print(find_squared_root(2))
print(find_squared_root(3))
```

```python
import numpy as np
import torch
import tensorflow as tf
```

```python
# Câu 1

arr_zeros_np = np.zeros((3, 4))
print('Dùng numpy :', arr_zeros_np)
print(' ')

arr_zeros_torch = torch.zeros((3, 4))
print('Dùng pytorch :', arr_zeros_torch)
print(' ')

arr_zeros_tensor = tf.zeros((3, 4))
print('Dùng tensor :', arr_zeros_tensor)
print(' ')
```

```python
# Câu 2

arr_ones_np = np.ones((3, 4))
print('Dùng numpy :', arr_ones_np)
print(' ')

arr_ones_torch = torch.ones((3, 4))
print('Dùng pytorch :', arr_ones_torch)
print(' ')

arr_ones_tensor = tf.ones((3, 4))
print('Dùng tensor :', arr_ones_tensor)
print(' ')
```

```python
# Câu 3

arr_np = np.full((3, 4), 5)
print('Dùng numpy :', arr_np)
print(' ')

arr_torch = torch.full((3, 4), 5)
print('Dùng pytorch :', arr_torch)
print(' ')

arr_tensor = tf.fill((3, 4), 5)
print('Dùng tensor :', arr_tensor)
print(' ')
```

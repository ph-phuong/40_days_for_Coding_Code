```python
corpus = ['Tôi thích môn Toán', 'Tôi thích AI', 'Tôi thích âm nhạc']

#nối các câu trong corpus thành một chuỗi duy nhất, sử dụng dấu cách làm phân cách
data = ' '.join(corpus)
data
```

```python
# dấu cách làm phân tách, tạo tập hợp các từ duy nhất trong danh sách đó, loại bỏ các từ trùng lặp
distinct_lst = list(set(data.split()))
distinct_lst
    
```

```python
sample = 'Tôi thích AI thích Toán'

#Phân tách chuỗi sample thành từng từ, cắt tại khoảng trắng của chuỗi
sample_word = sample.split()
sample_word
```

```python
#tạo 1 danh sách có số phần tử bằng phần tử trong distinct_lst, với giá trị là 0
sample_vector = [0]*len(distinct_lst)
sample_vector
```

```python
for word in sample_word:
    
    # Tìm vị trí của từ trong danh sách từ duy nhất (distinct_lst) bằng cách sử dụng phương thức index()
    word_index = distinct_lst.index(word)
    print(word_index)
    
    # Tăng giá trị của phần tử tương ứng trong sample_vector lên 1
    sample_vector[word_index] += 1
    
print(sample_vector)
```

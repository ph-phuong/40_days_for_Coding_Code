```python
corpus = ['Ti thch mn Ton', 'Ti thch AI', 'Ti thch m nhc']

#ni cc cu trong corpus thnh mt chui duy nht, s dng du cch lm phn cch
data = ' '.join(corpus)
data
```

```python
# du cch lm phn tch, to tp hp cc t duy nht trong danh sch , loi b cc t trng lp
distinct_lst = list(set(data.split()))
distinct_lst
    
```

```python
sample = 'Ti thch AI thch Ton'

#Phn tch chui sample thnh tng t, ct ti khong trng ca chui
sample_word = sample.split()
sample_word
```

```python
#to 1 danh sch c s phn t bng phn t trong distinct_lst, vi gi tr l 0
sample_vector = [0]*len(distinct_lst)
sample_vector
```

```python
for word in sample_word:
    
    # Tm v tr ca t trong danh sch t duy nht (distinct_lst) bng cch s dng phng thc index()
    word_index = distinct_lst.index(word)
    print(word_index)
    
    # Tng gi tr ca phn t tng ng trong sample_vector ln 1
    sample_vector[word_index] += 1
    
print(sample_vector)
```

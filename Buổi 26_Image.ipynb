```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# c v hin th nh
img1 = cv2.imread("Buoi_26-1.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
plt.axis('off')
plt.show()
```

```python
# c v hin th nh mu xm
img2 = cv2.imread("Buoi_26-1.jpg", 0)
plt.imshow(img2, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Tng cng  sng nh ln 50 n v
new_img1 = img1.astype(np.float32) + 50
new_img1 = np.clip(new_img1, 0, 255)
new_img1 = new_img1.astype(np.uint8)

plt.imshow(new_img1)
plt.axis('off')
plt.show()
```

```python
# Gim cng  sng nh ln 80 n v
new_img2 = img1.astype(np.float32) - 80
new_img2 = np.clip(new_img2, 0, 255)
new_img2 = new_img2.astype(np.uint8)

plt.imshow(new_img2)
plt.axis('off')
plt.show()
```

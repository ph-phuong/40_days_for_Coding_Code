```python
#Cu 1
with open('data_d21.txt', 'r') as f:
    data = f.read()
    data = data.replace('\n',' ').lower()
    print(data)
```

```python
# Cu 2
lst_data = data.split(' ')
distinct_word = list(set(lst_data))
print(sorted(distinct_word))
```

```python
# Cu 3
def count_word_text(lst, distinct_lst):
    count_word = {word : 0 for word in distinct_lst}
    for word in lst:
        if word in distinct_lst:
            count_word[word] += 1
    return count_word

word_counts = count_word_text(lst_data, distinct_word)
    
for word, count in word_counts.items():
    print(f'{word} in data is {count}')
    
max_word = max(word_counts, key=word_counts.get)
min_word = min(word_counts, key=word_counts.get)

print('\n')
print(f'{max_word} is a most frequent word : {word_counts[max_word]} times')
print(f'{min_word} is a least common word : {word_counts[min_word]} time')
```

```python
#Cu 3

data_vector = [0] * len(distinct_word)
for index, word in enumerate(distinct_word):
    data_vector[index] = lst_data.count(word)

print(f'{distinct_word} in data is {data_vector}')
```

```python
def calculate_can_chi_calender(year):
    list_can = ['Canh', 'Tn', 'Nhm', 'Qu', 'Gip', 't', 'Bnh', 'inh', 'Mu', 'K']
    list_chi = ['Thn', 'Du', 'Tut', 'Hi', 'T', 'Su', 'Dn', 'Mo', 'Thn', 'T', 'Ng', 'Mi']
    result = list_can[year%10] + " " + list_chi[year%12]
    return result

print('Nm 2024 l nm : ', calculate_can_chi_calender(2024))
print('Nm 1958 l nm : ', calculate_can_chi_calender(1958))
print('Nm 1959 l nm : ', calculate_can_chi_calender(1959))
print('Nm 1984 l nm : ', calculate_can_chi_calender(1984))
print('Nm 1987 l nm : ', calculate_can_chi_calender(1987))
print('Nm 1994 l nm : ', calculate_can_chi_calender(1994))
print('Nm 2019 l nm : ', calculate_can_chi_calender(2019))
```

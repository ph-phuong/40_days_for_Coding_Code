```python
def calculate_can_chi_calender(year):
    list_can = ['Canh', 'Tân', 'Nhâm', 'Quý', 'Giáp', 'Ất', 'Bính', 'Đinh', 'Mậu', 'Kỷ']
    list_chi = ['Thân', 'Dậu', 'Tuất', 'Hợi', 'Tý', 'Sửu', 'Dần', 'Mẹo', 'Thìn', 'Tỵ', 'Ngọ', 'Mùi']
    result = list_can[year%10] + " " + list_chi[year%12]
    return result

print('Năm 2024 là năm : ', calculate_can_chi_calender(2024))
print('Năm 1958 là năm : ', calculate_can_chi_calender(1958))
print('Năm 1959 là năm : ', calculate_can_chi_calender(1959))
print('Năm 1984 là năm : ', calculate_can_chi_calender(1984))
print('Năm 1987 là năm : ', calculate_can_chi_calender(1987))
print('Năm 1994 là năm : ', calculate_can_chi_calender(1994))
print('Năm 2019 là năm : ', calculate_can_chi_calender(2019))
```
